```python
# f=(c*1.8)+32
import tensorflow as tf
```

```python
import logging
logging.getLogger('tensorflow').setLevel(logging.ERROR)
```

```python
import numpy as np
```

```python
celcius=np.array([-40,-10,0,8,15,22,38], dtype=float)
fahrenheit=np.array([-40,14,32,46,59,72,100], dtype=float)
for i, c in enumerate(celcius):
  print("%.0f Celcius: %.0f Fahrenheit" % (c,fahrenheit[i]))
```

```python
x0=tf.keras.layers.Dense(units=1,input_shape=[1]) #creating the model
```

```python
model=tf.keras.Sequential([x0]) #model assembly
```

```python
model.compile (loss='mean_squared_error',
              optimizer=tf.keras.optimizers.Adam(0.1)) #compiling the model
```

```python
history=model.fit(celcius,fahrenheit,epochs=500,verbose=False)
print("finished training the model")
```

```python
import matplotlib.pyplot as plt
plt.xlabel('epoch number')
plt.ylabel('loss magnitude')
plt.plot(history.history['loss'])
```

```python
print(model.predict(np.array([100.0])))
```
